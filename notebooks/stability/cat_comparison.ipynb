```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib notebook
```

```python
length = 0.8 #m
width = 0.5 #m
draft = 0.15 #m
total_mass = length*width*draft*1000.0
print total_mass
KG = draft

g = 9.81
p = 1000.0

theta_max = np.arcsin(2*draft / width)
theta = np.linspace(0, theta_max, 1000)
```

```python
submerged_area = 0.5 * (2.0 * (draft / np.cos(theta) - width / 2.0 * np.tan(theta)) + width*np.tan(theta)) * width
h = draft / np.cos(theta) - width / 2.0 * np.tan(theta)
l = width * np.tan(theta)

M_mono = length * p * g * (-KG*submerged_area*np.sin(theta) + width*h**2*np.sin(theta)/2.0 + width*l*((h + l/3.0)*np.sin(theta) + width*np.cos(theta)/6.0)/2.0)
```

```python
b = width/2.0
print b
aspect = 4.0
M_cat = length * (2 * (aspect - 1.0)/2.0*b**2*np.sin(theta) * (aspect - 1.0)/2.0*b * p*g) + length * p * g * (-KG*submerged_area*np.sin(theta))
```

```python
plt.plot(np.degrees(theta), M_mono)
plt.plot(np.degrees(theta), M_cat)
plt.plot([17.4, 17.4], [0, 250], 'r--')
plt.xlabel('Roll Angle (degrees)')
plt.ylabel('Righting Moment (N-m)')
plt.legend(['Monohull','Catamaran'], loc=0)
plt.savefig('righting_moment.png')
plt.show()
```
