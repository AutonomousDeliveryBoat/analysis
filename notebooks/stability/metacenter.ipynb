```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib notebook
```

```python
catamaran_aspect_ratio = 4.0
T = np.linspace(0.1, 0.2, 1000)

length = 0.8
b = 60.0 / (1000.0*length*T) / 2.0
```

```python
KM_cat = (catamaran_aspect_ratio - 1.0)**2/4.0 * b**2 / T
KM_mono = b**2/(3*T) + T/2.0
```

```python
plt.xlabel('Draft (cm)')
plt.ylabel('Metacentric Height (m)')
plt.plot(T*100.0, KM_mono)
plt.plot(T*100.0, KM_cat)
plt.legend(['Monohull', 'Catamaran'])
plt.savefig('metacentric_height.png')
plt.show()
```
